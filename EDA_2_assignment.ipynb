**Both One-Hot Encoding `(OHE)` and Label Encoding `(LE)` are techniques for converting categorical data into numerical form for machine learning models. However, they have different use cases, advantages, and disadvantages.**

**one-hot encoding : Good for: Categories with no natural order (e.g., colors, cities)**  
**Label encoding : Good for: Ordered categories (e.g., Low, Medium, High)**

## **Feature Engineering:**

```python
df['income_per_hour'] = df['capital_gain'] / (df['hours_per_week'] )  # Avoid division by zero
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 35, 50, 65, 100], labels=['Teen', 'Young', 'Adult', 'Middle_Age', 'Senior'])
```

With the help of the above data we have created the two new columns they are the `income_per_hour`	`age_group`
`income_per_hour`  the  reason for creating this column is that the we can check the average income of a person  
`age_group` this is to describe about the age in the given columns in the form of the `'Teen', 'Young', 'Adult', 'Middle_Age', 'Senior'`

**Appilying the log transformaion to the captial_gain**

skewness befor transformation

```python
print("Skewness before:", df['capital_gain'].skew())
```

```python
df['capital_gain']=np.log(df['capital_gain'])
```

skewness after transformation

```python
print("Skewness after:", df['capital_gain'].skew())
```

"I applied a log transformation to the `capital_gain` feature because it was highly right-skewed, meaning most values were small while a few were extremely large. This transformation helped normalize the distribution, making it more suitable for models that assume normality, such as linear regression."
